```python
import pandas as pd

# Asegúrate de convertir las fechas a tipo datetime si no lo están
clientes['fechaingresocli'] = pd.to_datetime(clientes['fechaingresocli'])
ventas['fechaventa'] = pd.to_datetime(ventas['fechaventa'])

# Paso 1: Unir los DataFrames 'clientes' y 'ventas' por la columna 'idcliente' y 'idcli'
merged_df = pd.merge(ventas, clientes, left_on='idcli', right_on='idcliente', how='inner')

# Paso 2: Filtrar las ventas desde el 2023 en adelante
ventas_2023 = merged_df[merged_df['fechaventa'] >= '2023-01-01']

# Paso 3: Agrupar por 'idcliente' y obtener la última fecha de venta de cada cliente
ultimo_venta = ventas_2023.groupby('idcliente')['fechaventa'].max().reset_index()

# Paso 4: Filtrar los clientes activos de 2023
clientes_activos_2023 = ultimo_venta[ultimo_venta['fechaventa'] >= '2023-01-01']

# Unir con el DataFrame de clientes para obtener los detalles
clientes_activos_2023 = pd.merge(clientes_activos_2023, clientes, left_on='idcliente', right_on='idcliente', how='inner')
```

```python
import pandas as pd
import requests
from tqdm.notebook import tqdm

# Tu clave API de Google Maps
API_KEY = 'AIzaSyAUYMKwCM_MUD41vbPfklf-FVO-EGuEeaI'

# Filtrar solo filas con dirección válida y sin duplicados
cliente_concat = cliente_concat.dropna(subset=['direccion']).drop_duplicates(subset=['direccion']).copy()

# Función para consultar la API de Google Maps, con fallback a búsqueda internacional
def geocode(address):
    def consultar_api(direccion_completa):
        url = f'https://maps.googleapis.com/maps/api/geocode/json?address={direccion_completa.replace(" ", "+")}&key={API_KEY}'
        response = requests.get(url)
        return response.json()

    try:
        # Intento 1: buscar en Perú
        direccion_pe = f"{address}, Perú"
        result = consultar_api(direccion_pe)

        # Si falla, hacer segundo intento sin país
        if result['status'] != 'OK' or not result['results']:
            result = consultar_api(address)

        if result['status'] == 'OK':
            loc = result['results'][0]['geometry']['location']
            components = result['results'][0]['address_components']

            lat = loc['lat']
            lng = loc['lng']
            distrito = provincia = departamento = pais = None

            for comp in components:
                types = comp['types']
                if 'locality' in types:
                    distrito = comp['long_name']
                elif 'administrative_area_level_2' in types:
                    provincia = comp['long_name']
                elif 'administrative_area_level_1' in types:
                    departamento = comp['long_name']
                elif 'country' in types:
                    pais = comp['long_name']

            return pd.Series([lat, lng, distrito, provincia, departamento, pais])
        else:
            return pd.Series([None] * 6)
    except:
        return pd.Series([None] * 6)

# Aplicar con barra de progreso
tqdm.pandas()

# Crear el nuevo DataFrame 'cliente_ubicacion' con las columnas de ubicación
cliente_ubicacion = cliente_concat.copy()
cliente_ubicacion[['Latitud', 'Longitud', 'Distrito', 'Provincia', 'Departamento', 'Pais']] = cliente_ubicacion['direccion'].progress_apply(geocode)
```
